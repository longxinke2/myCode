```python
import pandas as pd
import numpy as np
import json
```

```python
data=pd.read_excel('C:/Users/Administrator/Desktop/简历表.xlsx')
data.head()
```

```python
data_list=pd.DataFrame(columns=[['name','sex','birth_place','marriage','birthday','mobile','email','city','job','job_category','salary',
                                 'education','degree','selfappraise','workpost','workstartday','workendday','workcontent','skill','certificate']])
for i in range(0,len(data)):
    a=json.loads(str(data['json'].loc[i]))
    name=a['Basic']['truename']
    sex=a['Basic']['sex']
    birth_place=a['Basic']['birth_place']
    if 'marriage' in a['Basic']:
        marriage=a['Basic']['marriage']
    else:
        marriage=[]
    if 'birthday' in a['Basic']:
        birthday=a['Basic']['birthday']
    else:
        birthday=[]
    if 'mobile' in a['Basic']:
        mobile=a['Basic']['mobile']
    else:
        mobile=[]
    if 'email' in a['Basic']:
        email=a['Basic']['email']
    else:
        email=[]
    if 'city' in a['Employment']:
        city=a['Employment']['city']
    else:
        city=[]
    if 'job' in a['Employment']:
        job=a['Employment']['job']
    else:
        job=[]
    if 'job_category' in a['Employment']:
        job_category=a['Employment']['job_category']
    else:
        job_category=[]
    if 'salary' in a['Employment']:
        salary=a['Employment']['salary']
    else:
        salary=[]
    if isinstance(a['Educational'],list) and len(a['Educational'])!=0:
        if 'tb_post' in a['Educational'][0]:
            education=a['Educational'][0]['tb_post']
        if 'tb_unittype' in a['Educational'][0]: 
            degree=a['Educational'][0]['tb_unittype']
    if isinstance(a['Selfappraise'],list) and len(a['Selfappraise'])!=0:
        if 'content' in a['Selfappraise'][0]:
            selfappraise=a['Selfappraise'][0]['content']
    else:
        selfappraise=[]
    if isinstance(a['Worked'],list) and len(a['Worked'])!=0:
        if 'tb_post' in a['Worked'][0]:
            workpost=a['Worked'][0]['tb_post']
        if 'tb_startday' in a['Worked'][0]:
            workstartday=a['Worked'][0]['tb_startday']
        if 'tb_endday' in a['Worked'][0]:
            workendday=a['Worked'][0]['tb_endday']
    else:
        workpost=[]
        workstartday=[]
        workendday=[]
    if isinstance(a['Worked'],list) and len(a['Worked'])!=0:
        if 'tb_content' in a['Worked'][0]:
            workcontent=a['Worked'][0]['tb_content']
    else:
        workcontent=[]
    if isinstance(a['Skills'],list) and len(a['Skills'])!=0:
        if 'text' in a['Skills'][0]:
            skill=a['Skills'][0]['text']
    else:
        skill=[]
    if isinstance(a['Certificate'],list) and len(a['Worked'])!=0 and len(a['Certificate'])!=0:
        if 'tb_unitname' in a['Certificate'][0]:
            certificate=list([a['Certificate'][0]['tb_unitname']])
        if len(a['Certificate'])==2:
            certificate=list([a['Certificate'][0]['tb_unitname'],a['Certificate'][1]['tb_unitname']])
        if len(a['Certificate'])==3:
            certificate=list([a['Certificate'][0]['tb_unitname'],a['Certificate'][1]['tb_unitname'],a['Certificate'][2]['tb_unitname']])
    else:
        certificate=[]
        
    data_list.loc[i]=[name,sex,birth_place,marriage,birthday,mobile,email,city,job,job_category,salary,education,degree,selfappraise,workpost,
                  workstartday,workendday,workcontent,skill,certificate]
    
```
