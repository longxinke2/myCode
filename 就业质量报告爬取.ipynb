```python
import requests
from lxml import etree
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
provinces = ['北京','天津','上海','重庆','黑龙江','吉林','辽宁','河北','河南','山东','山西','内蒙古','陕西','甘肃','宁夏','青海','新疆','江苏','江西','浙江','安徽','湖北','湖南','福建','广东','广西','海南','云南','四川','贵州','西藏']
first_chars = ['bj', 'tj', 'sh', 'cq', 'hlj', 'jl', 'ln', 'he', 'hn', 'sd', 'sx', 'nmg', 'sx', 'gs', 'nx', 'qh', 'xj', 'js', 'jx', 'zj', 'ah', 'hb', 'hn', 'fj', 'gd', 'gx', 'hi', 'yn', 'sc', 'gz', 'xz']
```

```python
# 拿每个省份的学校的网站

url_group = []
school_name = []

for i in range(0,len(provinces)):
    url = f'http://www.xiaoyuanzhaopin.net/{first_chars[i]}jyw/'

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    # 发送请求
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text,'html.parser')

    for i in soup.select("a[href]"):
        try:
            if '就业信息网' in i.string and i.string!='就业信息网':
                url_group.append(i['href'])
                school_name.append(re.sub('就业信息网','',i.string))
        except:
            pass
```

```python
for i in range(0,len(url_group)):
    url = url_group[i]
    if 'http://www.xiaoyuanzhaopin.net' in url:
        print(url)
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text,'html.parser')
        for j in soup.select("a[href]"):
            try:
                if 'cn' in j.string or 'com' in j.string:
                    url_group[i]=j['href']
                    break
            except:
                pass
```

```python
import re

def remove_domain_suffix(string):
    # 匹配".cn"或者".com"后面的部分
    pattern = re.compile(r'\.cn|\.com(/.*)?$')
    # 用顶级域名替换匹配到的部分
    result = re.sub(pattern, r'\1', string)
    return result
string = 'www.example.com/path/to/resource'
result = remove_domain_suffix(string)
print(result)  # 输出：www.example.com
```
