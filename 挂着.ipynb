```python
def click_image(image_path, confidence=0.9, lag_x=0,lag_y=0,wait=0):
    
    time.sleep(wait)
    # 从文件读取图像
    image = Image.open(image_path)

    # 全屏截图
    screen_image = ImageGrab.grab()

    # 在屏幕截图中查找匹配
    location = pyautogui.locate(image, screen_image, confidence=confidence)
    if location:
        # 匹配成功，返回位置信息
        x, y, a, b = location
        pyautogui.click(x+a//2+lag_x, y+b//2+lag_y)
        return 1
    else:
        # 匹配失败，返回None
        return 0

def qingjia():
    global flag
    if datetime.datetime.today().weekday() == 5 or datetime.datetime.today().weekday() == 6:
        return
    if flag == 1:
        return
    else:
        flag = 1
    confirm = pyautogui.confirm(text='请问是否开始请假?',title="提示")
    if confirm == 'Cancel':
        return
    for name in image_group:
        a = 0
        b = 0
        count = 0
        if name == 'rq.png':
            a = -10
        if name in['kssj.png','jssj.png','qjlx1.png','gzdlr.png']:
            a = 190
        if name in ['qjlx1.png','gzdlr.png']:
            time.sleep(1)# 加载的时候位置会浮动
        while click_image(image_path+name,lag_x=a,lag_y=b)==0:
            time.sleep(0.2)
            count +=1
            if count > 10:#触发意外检测
                while True:
                    click_image(image_path + 'qywxh.png')
                    click_image(image_path + 'spc.png')
                    time.sleep(1)
    #时间选择单独处理
    #开始时间
    click_image(image_path+'kssj.png',lag_x=400,wait=0.5)
    click_image(image_path+'qd.png',lag_x=-10,lag_y=-40,wait=0.5)

    click_image(image_path+'qd.png',lag_x=-10,lag_y=-40,wait=0.5)
    click_image(image_path+'qd.png',lag_x=-10,lag_y=-40,wait=0.5)
    click_image(image_path+'qd.png',lag_x=-10,lag_y=-40,wait=0.5)
    click_image(image_path+'qd.png',lag_x=-10,lag_y=-250,wait=0.5)
    click_image(image_path+'qd.png',lag_x=55,lag_y=-300,wait=0.5)
    click_image(image_path+'qd.png',lag_x=55,lag_y=-300,wait=0.5)
    click_image(image_path+'qd.png',lag_x=55,lag_y=-300,wait=0.5)
    click_image(image_path+'qd.png',lag_x=55,lag_y=-300,wait=0.5)
    click_image(image_path+'qd.png',wait=0.5)
    #结束时间
    click_image(image_path+'jssj.png',lag_x=400,wait=0.5)
    click_image(image_path+'qd.png',lag_x=-10,lag_y=-40,wait=0.5)
    click_image(image_path+'qd.png',lag_x=-10,lag_y=-40,wait=0.5)
    click_image(image_path+'qd.png',lag_x=-10,lag_y=-40,wait=0.5)
    click_image(image_path+'qd.png',lag_x=-10,lag_y=-40,wait=0.5)
    click_image(image_path+'qd.png',lag_x=-10,lag_y=-250,wait=0.5)
    click_image(image_path+'18.png',wait=0.5)

    click_image(image_path+'qd.png',lag_x=55,lag_y=-300,wait=0.5)
    click_image(image_path+'qd.png',lag_x=55,lag_y=-300,wait=0.5)
    click_image(image_path+'qd.png',lag_x=55,lag_y=-300,wait=0.5)
    click_image(image_path+'qd.png',lag_x=55,lag_y=-300,wait=0.5)
    click_image(image_path+'qd.png',wait=0.5)

    click_image(image_path+'tj.png',wait=0.5)
    time.sleep(60)
    flag = 0
```
