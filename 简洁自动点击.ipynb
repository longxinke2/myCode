```python
import pyautogui
import shutil
import time
from PIL import ImageGrab
from PIL import Image
import os
import keyboard
import pyperclip
import wx
import threading
import lxk

image_path = 'D:/AutoClickPicture/'
cache = ''
thread_flag=False
try:
    os.mkdir(image_path)
except:
    pass
def click_image(image, confidence=0.95, lag_x=0,lag_y=0,wait=0.1, drag=None, double=False):
    try:
        while True:
            time.sleep(wait)
            # 获取屏幕分辨率
            screen_width, screen_height = pyautogui.size()

            # 全屏截图
            screen_image = ImageGrab.grab()

            # 在屏幕截图中查找匹配
            location = pyautogui.locate(image, screen_image, confidence=confidence)
            if location:
                # 匹配成功，返回位置信息
                x, y, a, b = location
                x1,y1=pyautogui.position()
                if drag==None:
                    if double==False:
                        pyautogui.click(x+a//2+lag_x, y+b//2+lag_y)
                    else:
                        pyautogui.doubleClick(x+a//2+lag_x, y+b//2+lag_y)
                else:
                    pyautogui.moveTo(x+a//2+lag_x, y+b//2+lag_y)
                    pyautogui.mouseDown()
                    pyautogui.moveRel(drag[0],drag[1],0.2)
                    pyautogui.mouseUp()
                return (x1,y1)
            else:
                # 匹配失败，返回None
                if thread_flag==False:return False
                
    except Exception as e:
        print(e)
```

```python
class btnFrame(wx.Frame):
    def __init__( self, parent ):
        self.wait = 0
        self.mouse=[500,500]
        self.move = [0,0]
        self.flag = -1
        self.loop=False
        self.temp=''
        self.pic_group =[]
        
        self.thread1=threading.Thread(target=self.stop_byf2)
#         self.thread1.setDaemon(True)
        self.thread1.start()
        
        wx.Frame.__init__ ( self, parent, id = wx.ID_ANY, title = '自动点击程序', pos = wx.DefaultPosition, size = wx.Size( 269,310 ), style = wx.DEFAULT_FRAME_STYLE|wx.TAB_TRAVERSAL )

        self.SetSizeHints( wx.DefaultSize, wx.DefaultSize )

        bSizer1 = wx.BoxSizer( wx.HORIZONTAL )

        bSizer2 = wx.BoxSizer( wx.VERTICAL )

        self.m_button1 = wx.Button( self, wx.ID_ANY, u"添加图片", wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer2.Add( self.m_button1, 0, wx.ALL, 5 )

        self.m_staticText1 = wx.StaticText( self, wx.ID_ANY, u"文件夹名称", wx.DefaultPosition, wx.DefaultSize, 0 )
        self.m_staticText1.Wrap( -1 )

        bSizer2.Add( self.m_staticText1, 0, wx.ALL, 5 )

        self.m_textCtrl1 = wx.TextCtrl( self, wx.ID_ANY, wx.EmptyString, wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer2.Add( self.m_textCtrl1, 0, wx.ALL, 5 )

        self.m_button2 = wx.Button( self, wx.ID_ANY, u"保存", wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer2.Add( self.m_button2, 0, wx.ALL, 5 )

        self.m_button3 = wx.Button( self, wx.ID_ANY, u"加载", wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer2.Add( self.m_button3, 0, wx.ALL, 5 )

        self.m_button4 = wx.Button( self, wx.ID_ANY, u"执行", wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer2.Add( self.m_button4, 0, wx.ALL, 5 )

        self.m_button5 = wx.Button( self, wx.ID_ANY, u"循环执行", wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer2.Add( self.m_button5, 0, wx.ALL, 5 )

        self.m_staticText2 = wx.StaticText( self, wx.ID_ANY, u"停止按F2", wx.DefaultPosition, wx.DefaultSize, 0 )
        self.m_staticText2.Wrap( -1 )

        bSizer2.Add( self.m_staticText2, 0, wx.ALL, 5 )


        bSizer1.Add( bSizer2, 1, wx.EXPAND, 5 )

        bSizer3 = wx.BoxSizer( wx.VERTICAL )

        self.m_staticText3 = wx.StaticText( self, wx.ID_ANY, u"运行信息", wx.DefaultPosition, wx.DefaultSize, 0 )
        self.m_staticText3.Wrap( -1 )

        bSizer3.Add( self.m_staticText3, 0, wx.ALL, 5 )

        self.m_textCtrl2 = wx.TextCtrl( self, wx.ID_ANY, wx.EmptyString, wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer3.Add( self.m_textCtrl2, 0, wx.ALL, 5 )

        self.m_button6 = wx.Button( self, wx.ID_ANY, u"始终置顶", wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer3.Add( self.m_button6, 0, wx.ALL, 5 )

        self.m_button7 = wx.Button( self, wx.ID_ANY, u"多等待0.5s", wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer3.Add( self.m_button7, 0, wx.ALL, 5 )

        self.m_button8 = wx.Button( self, wx.ID_ANY, u"设置鼠标位移", wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer3.Add( self.m_button8, 0, wx.ALL, 5 )

        bSizer4 = wx.BoxSizer( wx.HORIZONTAL )

        self.m_staticText4 = wx.StaticText( self, wx.ID_ANY, u"x", wx.DefaultPosition, wx.DefaultSize, 0 )
        self.m_staticText4.Wrap( -1 )

        bSizer4.Add( self.m_staticText4, 0, wx.ALL, 5 )

        self.m_textCtrl3 = wx.TextCtrl( self, wx.ID_ANY, wx.EmptyString, wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer4.Add( self.m_textCtrl3, 0, wx.ALL, 5 )


        bSizer3.Add( bSizer4, 1, wx.EXPAND, 5 )

        bSizer5 = wx.BoxSizer( wx.HORIZONTAL )

        self.m_staticText5 = wx.StaticText( self, wx.ID_ANY, u"y", wx.DefaultPosition, wx.DefaultSize, 0 )
        self.m_staticText5.Wrap( -1 )

        bSizer5.Add( self.m_staticText5, 0, wx.ALL, 5 )

        self.m_textCtrl4 = wx.TextCtrl( self, wx.ID_ANY, wx.EmptyString, wx.DefaultPosition, wx.DefaultSize, 0 )
        bSizer5.Add( self.m_textCtrl4, 0, wx.ALL, 5 )


        bSizer3.Add( bSizer5, 1, wx.EXPAND, 5 )


        bSizer1.Add( bSizer3, 1, wx.EXPAND, 5 )


        self.SetSizer( bSizer1 )
        self.Layout()

        self.Centre( wx.BOTH )
        self.Show()
        self.SetWindowStyle(wx.DEFAULT_FRAME_STYLE | wx.STAY_ON_TOP)

    
        self.m_button1.Bind(wx.EVT_BUTTON,self.On_click_btn)
        self.m_button2.Bind(wx.EVT_BUTTON, self.On_save_btn)
        self.m_button3.Bind(wx.EVT_BUTTON, self.On_load_btn)
        self.m_button4.Bind(wx.EVT_BUTTON, self.On_start_btn)
        self.m_button5.Bind(wx.EVT_BUTTON, self.On_loop_btn)
        self.m_button6.Bind(wx.EVT_BUTTON, self.On_toggle_btn)
        self.m_button7.Bind(wx.EVT_BUTTON, self.On_wait_btn)
        self.m_button8.Bind(wx.EVT_BUTTON, self.On_move_btn)
    
    def On_wait_btn(self, event):
        self.wait = 0.5+self.wait
        self.m_textCtrl2.SetValue(str(self.wait))
        
    def On_move_btn(self, event):
        self.move = [0 if self.m_textCtrl3.GetValue()=='' else int(self.m_textCtrl3.GetValue()),0 if self.m_textCtrl4.GetValue()=='' else int(self.m_textCtrl4.GetValue())]
        
    def On_toggle_btn(self, event):
        if self.GetWindowStyle() & wx.STAY_ON_TOP:
            self.SetWindowStyle(wx.DEFAULT_FRAME_STYLE)
            self.m_textCtrl2.SetValue('取消置顶')
        else:
            self.SetWindowStyle(wx.DEFAULT_FRAME_STYLE | wx.STAY_ON_TOP)
            self.m_textCtrl2.SetValue('已置顶')
            
    def On_click_btn(self, event):
        self.flag +=1
        self.pic_group.insert(self.flag,ImageGrab.grabclipboard())
        if click_image(self.pic_group[self.flag],wait=self.wait):
            self.m_textCtrl2.SetValue('点击成功')
        else:
            self.m_textCtrl2.SetValue('点击失败')
    
    def On_debug_btn(self, event):
        x,y=pyautogui.position()
        if self.flag==len(self.pic_group)-1:return
        self.flag+=1
        if click_image(self.pic_group[self.flag],wait=self.wait):
            self.m_textCtrl2.SetValue('点击成功')
        else:
            self.m_textCtrl2.SetValue('点击失败')  
        pyautogui.moveTo(x,y)
        
    def On_reset_btn(self, event):
        self.flag=-1
    
    def On_save_btn(self, event):
        global cache
        self.temp = self.m_textCtrl1.GetValue()
        cache = self.temp
        try:
            os.mkdir(image_path+f'{self.temp}/')
        except:
            shutil.rmtree(image_path+f'{self.temp}/')
            os.mkdir(image_path+f'{self.temp}/')
        for i in range(0,len(self.pic_group)):
            self.pic_group[i].save(image_path+f"{self.temp}/{'0'+str(i) if i<10 else i}.png")
        
    def On_load_btn(self, event):
        global cache
        cache = self.temp
        self.temp = self.m_textCtrl1.GetValue()
        self.pic_group=[]
        self.flag=-1
        for i in os.listdir(image_path+f'{self.temp}/'):
            image = Image.open(image_path+f'{self.temp}/{i}')
            self.pic_group.append(image)
    
    def start_thread(self):
        global thread_flag
        l=len(self.pic_group)
        while True:
            for i in range(l):
                if thread_flag==True:
                    temp = click_image(self.pic_group[i])
                    if temp and self.loop and i==0:
                        self.mouse[0],self.mouse[1]=temp
                    self.m_textCtrl2.SetValue(lxk.progress_bar(i+1,l,'',10))
                else:
                    break
            if not self.loop:break
            pyautogui.moveTo(self.mouse[0]+self.move[0],self.mouse[1]+self.move[1])
        thread_flag=False
        self.m_textCtrl2.SetValue('stop')
        
    def stop_byf2(self):
        while True:
            keyboard.wait('f2')
            global thread_flag
            if thread_flag:#关闭线程
                thread_flag=False
                self.loop=False
                self.flag=-1
                self.m_textCtrl2.SetValue('stop')
                while self.loop:
                    thread_flag=False
                    self.loop=False
                    self.flag=-1
                    self.m_textCtrl2.SetValue('stop')
        
    def On_start_btn(self, event):
        global thread_flag
        thread_flag=True
        self.thread=threading.Thread(target=self.start_thread)
        self.thread.start()
        
    def On_loop_btn(self, event):
        self.loop=True
        self.On_start_btn(event)
        
    
app = wx.App()
frame = btnFrame(None)
app.MainLoop()
try:
    del app
except:
    pass
```
