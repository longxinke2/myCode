# 合并单元格拆分

```python
import pandas as pd
import numpy as np
path='E:/桌面/2022湖北省单位行业.xlsx'
data = pd.read_excel(path)
```

```python
#设置需要合并的列
group=[0,1,2]
for j in group:
    for i in range(data.shape[0]):
        if pd.isna(data.iloc[i,j]):
            data.iloc[i,j]=data.iloc[i-1,j]

data.to_excel(path)
```
