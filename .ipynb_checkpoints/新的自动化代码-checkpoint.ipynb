```python
import wx

class Frame(wx.Frame):
    def __init__(self, parent, title):
        super(Frame, self).__init__(parent, title=title, size=(300, 200))
        panel = wx.Panel(self)
        vbox = wx.BoxSizer(wx.VERTICAL)
        hbox = wx.BoxSizer(wx.HORIZONTAL)
        
        # 创建下拉框并添加选项
        label = wx.StaticText(panel, label="选择一项:")
        self.combo_box = wx.ComboBox(panel, choices=["a", "b", "c"])
        hbox.Add(label, wx.EXPAND | wx.ALL, 5)
        hbox.Add(self.combo_box, wx.EXPAND | wx.ALL, 5)
        vbox.Add(hbox, wx.EXPAND | wx.ALL, 5)

        # 创建一个文本框
        self.text_ctrl = wx.TextCtrl(panel, style=wx.TE_MULTILINE)
        self.text_ctrl.Hide()   # 首先将文本框隐藏

        # 注册下拉框选项事件
        self.combo_box.Bind(wx.EVT_COMBOBOX, self.on_select)

        vbox.Add(self.text_ctrl, wx.EXPAND | wx.ALL, 5)
        panel.SetSizer(vbox)
        self.Show()

    def on_select(self, event):
        selection = self.combo_box.GetStringSelection()
        if selection == "a":
            self.text_ctrl.Show()
        else:
            self.text_ctrl.Hide()

        self.Layout()

if __name__ == '__main__':
    app = wx.App()
    frame = Frame(None, "下拉框示例")
    app.MainLoop()
```
